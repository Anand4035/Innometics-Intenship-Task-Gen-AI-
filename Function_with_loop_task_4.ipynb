1.Mobile Recharge Validation System

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount < 50:
        return "Recharge amount must be at least ₹50."
    elif amount not in valid_plans:
        return f"Invalid plan! Available plans are: {valid_plans}"
    else:
        return "Recharge Successful! "
# Retry system using while loop
while True:
    try:
        user_amount = int(input("Enter recharge amount: ₹"))
        result = validate_recharge(user_amount)
        print(result)
        if result == "Recharge Successful! ":
            break  # Exit loop if valid
    except ValueError:
        print("Please enter a valid numeric amount.")
```

 2.Inventory Reorder Alert System


```python
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert  (Stock: {quantity})")
        else:
            print(f"{product}: Stock OK  (Stock: {quantity})")
# Sample Inventory Data
inventory = {
    "Laptop": 20,
    "Mouse": 10,
    "Keyboard": 15,
    "Monitor": 8,
    "USB Cable": 25
}
# Function Call
check_inventory(inventory)
```

3.Student Result Processing System


```python
def process_result(marks):
    total = 0
    # Loop to calculate total marks
    for mark in marks:
        total += mark
    # Calculate average
    average = total / len(marks)
    # Determine result
    if average >= 50:
        return f"Average Marks: {average:.2f} → Pass "
    else:
        return f"Average Marks: {average:.2f} → Fail "
# Sample Data
student_marks = [60, 45, 70, 55, 40]
# Function Call
result = process_result(student_marks)
print(result)
```

4.Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, is_peak):
    base_fare = 50
    per_km_rate = 12
    # Calculate normal fare
    fare = base_fare + (distance * per_km_rate)
    # Add 25% extra if peak hour
    if is_peak:
        fare += fare * 0.25
    return fare
# Retry option using while loop
while True:
    try:
        distance = float(input("Enter distance (in km): "))
        peak_input = input("Is it peak hour? (yes/no): ").lower()
        if peak_input == "yes":
            is_peak = True
        else:
            is_peak = False
        total_fare = calculate_fare(distance, is_peak)
        print(f"Total Cab Fare: ₹{total_fare:.2f} ")
        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            print("Thank you for using Cab Fare Estimator!")
            break
    except ValueError:
        print("Please enter valid numeric distance.")
```

 5.Employee Attendance Eligibility Checker


```python
def check_attendance(attendance_list):
    present_count = 0
    # Loop to count present days
    for status in attendance_list:
        if status == "P":
           present_count += 1
    # Calculate attendance percentage
    total_days = len(attendance_list)
    percentage = (present_count / total_days) * 100
    # Determine eligibility
    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% → Eligible "
    else:
        return f"Attendance: {percentage:.2f}% → Not Eligible "
# Sample Data
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
# Function Call
result = check_attendance(attendance)
print(result)
```

 6.Password Strength Checker


```python
def check_password_strength(password):
    has_digit = False
    has_special = False
    # Check minimum length
    if len(password) < 8:
        return "Weak Password  (Minimum 8 characters required)"
    # Loop through each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True
    # Check conditions
    if has_digit and has_special:
        return "Strong Password "
    else:
        return "Weak Password  (Must include at least one digit and one special character @#$)"
# Example Usage
user_password = input("Enter your password: ")
result = check_password_strength(user_password)
print(result)
```
