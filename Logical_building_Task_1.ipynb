2.Pass/Fail Analyzer

```python
marks=[45,78,90,33,60]
pass_students = 0
fail_students = 0
for mark in marks:
  if mark >= 50:
    pass_students += 1
  else:
    fail_students += 1
print("Total Students:", len(marks))
print("The total number of pass students:", pass_students)
print("The total number of fail students:", fail_students)
```

3.Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names=[]
for name in names:
  cleaned_name=name.strip().lower()
  cleaned_names.append(cleaned_name)
print("Cleaned names:",cleaned_names)
```

4.Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    length=len(msg)
    print("Message:", msg)
    print("Length:", length)
    if length > 10:
        print("Flag: Message longer than 10 characters")
```

5.Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_counts = 0
for log in logs:
    if log == "ERROR":
        error_counts+= 1
print("Total ERROR entries:", error_counts)
```
