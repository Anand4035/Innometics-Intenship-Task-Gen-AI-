1.Unique word in a sentence


```python
sentence=input("Enter the sentence:")
se_spl=sentence.split()
remove_dup=set(se_spl)
print(len(remove_dup))
print(remove_dup)
```

2.Highest Salary from employee data


```python
employees={"Ravi":75000,"Ankita":68000,"Kiran":72000}
highest_employee=" "
highest_salary=0
for name,salary in employees.items():
  if salary>highest_salary:
    highest_salary=salary
    highest_employee=name
print("Highest Salary:",highest_employee,"-",highest_salary)
```

3.Find maximum and minimum values

```python
numbers = [45, 22, 89, 10, 66]
maximum = numbers[0]
minimum = numbers[0]
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

4.Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]
count = len([price for price in prices if price > 1000])
print("Products above 1000:", count)
```

5.Calculate Attendence Percentage

```python
attendance = ["P", "P", "A", "P", "P"]
present_count = 0
total_days = len(attendance)
for day in attendance:
    if day == "P":
        present_count += 1
percentage = (present_count / total_days) * 100
print("Attendance Percentage:", percentage)
```

6.Remove Duplicate Phone Numbers

```python
phone_numbers = [9876543210, 9123456789, 9876543210]
unique_numbers = set(phone_numbers)
print("Unique phone numbers:", unique_numbers)
```

7.Count Character Frequency

```python
text = "python"
char_count = {}
for char in text:
    char_count[char] = text.count(char)
print(char_count)
```
