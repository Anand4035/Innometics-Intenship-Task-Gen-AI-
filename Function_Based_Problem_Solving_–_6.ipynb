1.Smart Parking Lot Management System

```python
def smart_parking(capacity, logs):
    current = 0
    peak = 0
    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            current -= 1
        if current > peak:
            peak = current
    print("Currently Parked Vehicles:", current)
    if current > capacity:
        print("Parking Status: Full - Capacity Exceeded")
    else:
        print("Parking Status: Available")
    print("Peak Usage:", peak)
# Sample Run
smart_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

2.Online Food Delivery Time Estimator

```python
def estimate_delivery(distance, traffic, weather):
    base_time = distance * 5   # 5 minutes per km
    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 10
    if weather == "Rainy":
        base_time += 10
    elif weather == "Stormy":
        base_time += 20
    print("Estimated Delivery Time:", base_time, "minutes")
estimate_delivery(8, "High", "Rainy")
```

3.Movie Theatre Seat Occupancy Analyzer

```python
def theatre_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    percentage = (booked_count / total_seats) * 100
    print("Occupancy:", str(int(percentage)) + "%")
    if percentage == 100:
        print("Show Status: Housefull")
    elif percentage >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")
    if percentage > 90:
        print("Suggestion: Consider opening additional shows")
theatre_occupancy(200, [1]*150)
```

4: Cloud Server Load Classification System

```python
def server_load(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading
    average = total / len(cpu_readings)
    print("Average CPU Load:", int(average), "%")
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
server_load([45, 60, 70, 85, 90])
```

5.Smart Classroom Resource Usage Monitor

```python
def classroom_monitor(resources):
    overused = []
    alert = "No"
    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)
            alert = "Yes"
    if overused:
        print("Overused Resources:", ", ".join(overused))
    else:
        print("Overused Resources: None")
    print("Energy Alert:", alert)
classroom_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

6.Online Event Registration Capacity Controller

```python
def event_registration(capacity, registrations):
    confirmed = 0
    waitlisted = 0
    for _ in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    if waitlisted > 0:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")
event_registration(100, 105)
```
