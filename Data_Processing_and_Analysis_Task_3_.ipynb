1.Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
# Find highest performance score
highest_score = max(employees.values())
# Find employees with highest score
top_performers = [name for name, score in employees.items() if score == highest_score]
# Display result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

2.Search Query Keyword Analysis

```python
import string
query = "Buy mobile phone buy phone online"
# Convert to lowercase
query = query.lower()
# Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))
# Split into words
words = query.split()
# Count frequency of each word
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1
# Display words searched more than once
result = {word: count for word, count in word_count.items() if count > 1}
print(result)
```

3.Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
# Store valid readings (even numbers) as (hour_index, value)
valid_readings = [(index, value)
                  for index, value in enumerate(sensor_readings)
                  if value % 2 == 0]
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

4.Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
# Count domains
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1
# Total emails
total_users = len(emails)
# Calculate and display percentage usage
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {int(percentage)}%")
```

5.Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
# Calculate average sales
average_sales = sum(sales) / len(sales)
# Threshold for spike (30% above average)
threshold = average_sales * 1.3
# Detect spike days
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

6.Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
# Dictionary to store frequency of user IDs
user_count = {}
# Count occurrences of each user ID
for user in user_ids:
    user_count[user] = user_count.get(user, 0) + 1
# Display duplicate user IDs and their counts
for user, count in user_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
